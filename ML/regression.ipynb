Handling missing values

Techniques:

Drop rows/columns

Fill numeric values with mean/median

Fill categorical values with mode

Use advanced imputers (KNN, IterativeImputer)

```python
#Fill numeric values with mean/median
#Fill categorical values with mode
#df.fillna(df.mean(), inplace=True)

df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].median())
```

```python
df['total_bedrooms'].value_counts()
```

```python
df.head()
```

```python
## simple imputer
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
imputer.fit(df[["total_bedrooms"]])
df[["total_bedrooms"]] = imputer.transform(df[["total_bedrooms"]])
```

```python
cat_imputer = SimpleImputer(strategy="most_frequent")
cat_imputer.fit(df[["ocean_proximity"]])
df[["ocean_proximity"]] = cat_imputer.transform(df[["ocean_proximity"]])
```
